## Tip 4 - Provide an Python SDK

```python

from swcpy import SWCClient, SWCConfig
import pandas as pd
import os

api_base_url = os.getenv('API_BASE_URL')
config = SWCConfig(url=api_base_url,backoff=False)
client = SWCClient(config)

#get all players from API
sdk_players = client.list_players()

#convert the list of players instances to dictionaries
df = pd.DataFrame([player.model_dump() for player in sdk_players])
display(df)
```

# Tip 5 - Provide Bulk Files with SDK

```python
#get a list of players from bulk file
player_file = client.get_bulk_player_file()

#write the file to disk to verify file download
output_file_path = 'players_file.csv' 
with open(output_file_path, 'wb') as f:
    f.write(player_file)
```

# Tip  6 - Last Changed Date with SDK

```python
#get players from sdk with last changed date
sdk_players = client.list_players(minimum_last_changed_date="2024-09-15")

#convert the list of players instances to dictionaries
df = pd.DataFrame([player.model_dump() for player in sdk_players])
display(df)
```
